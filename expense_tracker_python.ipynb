expense-tracker-python

```python
from collections import defaultdict
from datetime import datetime

expenses = []

def add_expense(amount, category, description):
    expenses.append({
        "date": datetime.now().strftime("%Y-%m-%d"),
        "amount": amount,
        "category": category,
        "desc": description
    })

def summary():
    cat_totals = defaultdict(float)
    for e in expenses:
        cat_totals[e['category']] += e['amount']
    total = sum(cat_totals.values())
    print("\n===== EXPENSE SUMMARY =====")
    for cat, amt in sorted(cat_totals.items(), key=lambda x: -x[1]):
        pct = (amt / total) * 100
        bar = '█' * int(pct / 5)
        print(f"{cat:<15} ₹{amt:>8.2f}  {bar} {pct:.1f}%")
    print(f"{'TOTAL':<15} ₹{total:>8.2f}")
    biggest = max(cat_totals, key=cat_totals.get)
    print(f"\nBiggest spending: {biggest} (₹{cat_totals[biggest]:.2f})")

add_expense(500, "Food", "Lunch at restaurant")
add_expense(1200, "Transport", "Monthly metro pass")
add_expense(800, "Shopping", "New shirt")
add_expense(300, "Food", "Groceries")
add_expense(2000, "Bills", "Electricity")
add_expense(450, "Entertainment", "Movie tickets")
summary()
```
